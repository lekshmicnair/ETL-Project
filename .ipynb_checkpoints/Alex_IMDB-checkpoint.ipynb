```python
import pandas as pd
```

```python
imdb_raw = pd.read_csv("Resources/movies.csv", encoding='latin-1')
imdb_raw
```

```python
IMDB_Vote_Count = imdb_raw[['name', 'votes']]
IMDB_Vote_Count.rename(columns = {'name': 'Movie_title', 'votes': 'Vote_Count'}, inplace=True)
IMDB_Vote_Count.set_index('Movie_title', inplace=True)
IMDB_Vote_Count
```

```python
IMDB_Score = imdb_raw[['name', 'score']]
IMDB_Score.rename(columns = {'name': 'Movie_title', 'score': 'IMDB_Score'}, inplace=True)
IMDB_Score.set_index('Movie_title', inplace=True)
IMDB_Score
```

```python
IMDB_Data = imdb_raw[['name', 'budget', 'gross', 'country', 'company', 'director', 'star', 'rating']]
IMDB_Data.rename(columns = {'name': 'Movie_title', 'budget': 'Budget', 'gross': 'Gross', 'country': 'Country', 'company': 'Company', 'director': 'Director', 'star': 'Star', 'rating': 'Rating'}, inplace=True)
IMDB_Data.set_index('Movie_title', inplace=True)
IMDB_Data
```
